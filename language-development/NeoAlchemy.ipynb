```python
class Comparator(object):
    def rshift(self, a, b):
        raise NotImplementedError
    def lshift(self, a, b):
        raise NotImplementedError
    def sub(self, a, b):
        raise NotImplementedError


class Element(object):
    type_ = 'element'
    comparator_factory = Comparator
    
    def render(self, **ctx):
        return ''
    
    def __repr__(self):
        return self.render()
    
    def __unicode__(self):
        return self.render()
    
    def __rshift__(self, other):
        return self.comparator_factory().rshift(self, other)
    
    def __lshift__(self, other):
        return self.comparator_factory().lshift(self, other)
    
    def __sub__(self, other):
        return self.comparator_factory().sub(self, other)


class NodeLike(Comparator):
    def rshift(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            ghost_rel = Connector(
                AnonymousRelationship(), b, Connector.RIGHT)
            return Connector(a, ghost_rel, Connector.UNDIRECTED)
        else:
            return Connector(a, b, Connector.RIGHT)

    def lshift(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            ghost_rel = Connector(
                AnonymousRelationship(), b, Connector.UNDIRECTED)
            return Connector(a, ghost_rel, Connector.LEFT)
        else:
            return Connector(a, b, Connector.LEFT)
        
    def sub(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            ghost_rel = Connector(
                AnonymousRelationship(), b, Connector.UNDIRECTED)
            return Connector(a, ghost_rel, Connector.UNDIRECTED)
        else:
            return Connector(a, b, Connector.UNDIRECTED)


class Node(Element):
    creation_counter = 1
    comparator_factory = NodeLike
    type_ = 'node'
    prefix = '('
    suffix = ')'
    
    def __init__(self, *labels, **props):
        self.idx = Node.creation_counter
        Node.creation_counter += 1
        self.variable = '%s_%d' % (self.type_, self.idx)
        self.labels = labels
        self.props = props
    
    def render(self, **ctx):
        parts = []
        if self.variable:
            parts.append(self.variable)
        if self.labels:
            parts.append(':' + ':'.join(self.labels))
        if self.props:
            if parts:
                parts.append(' ')
            parts.append(repr(self.props))  # don't do this in real life, yo
        return ''.join((self.prefix, ''.join(parts), self.suffix))


class Relationship(Node):
    type_ = 'rel'
    prefix = '['
    suffix = ']'


class AnonymousRelationship(Relationship):
    def __init__(self, *args, **kwargs):
        super(AnonymousRelationship, self).__init__(*args, **kwargs)
        self.variable = ''


class Connector(Element):
    UNDIRECTED = 0
    LEFT = 1
    RIGHT = 2
    
    type_ = 'connector'
    comparator_factory = NodeLike
    
    def __init__(self, left, right, direction):
        self.left = left
        self.right = right
        self.direction = direction
    
    def render(self, **ctx):
        connector = {
            self.UNDIRECTED: '-',
            self.LEFT: '<-',
            self.RIGHT: '->',
        }[self.direction]
        return ''.join([self.left.render(**ctx), connector, 
                        self.right.render(**ctx)])


print Node('Ingredient', name='Soy Milk') - Relationship('EXCLUDED_FROM') >> Node('Ingredient', name='Milk')
```
