```python
class Comparator(object):
    def rshift(self, a, b):
        raise NotImplementedError
    def lshift(self, a, b):
        raise NotImplementedError
    def sub(self, a, b):
        raise NotImplementedError


class Element(object):
    type_ = 'element'
    comparator_factory = Comparator
    
    def render(self, **ctx):
        return ''
    
    def match_expression(self):
        return None
    
    def returnable(self):
        return None
    
    def __repr__(self):
        return self.render()
    
    def __unicode__(self):
        return self.render()
    
    def __rshift__(self, other):
        return self.comparator_factory().rshift(self, other)
    
    def __lshift__(self, other):
        return self.comparator_factory().lshift(self, other)
    
    def __sub__(self, other):
        return self.comparator_factory().sub(self, other)


class Text(Element):
    type_ = 'text'
    
    def __init__(self, text):
        self.text = text
    
    def render(self, **ctx):
        return self.text
    
    def returnable(self):
        return self


class Empty(Text):
    type_ = 'empty'
    
    def __init__(self):
        super(Empty, self).__init__('')
    
    def returnable(self):
        return None


class NodeLikeComparator(Comparator):
    def rshift(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            b = Connector(
                AnonymousRelationship(), b, Connector.RIGHT)
            ret = lambda: Connector(a, b, Connector.UNDIRECTED)
        else:
            ret = lambda: Connector(a, b, Connector.RIGHT)
            
        if a.type_ == 'connector' and a.right.type_ == 'relationship' and a.direction != Connector.UNDIRECTED:
            raise Exception('Other side of a relationship must be undirected.\n'
                            'Use - instead of >>: Node() - Relationship() >> Node()')
        
        return ret()

    def lshift(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            b = Connector(
                AnonymousRelationship(), b, Connector.UNDIRECTED)
        
        if a.type_ == 'connector' and a.right.type_ == 'relationship' and a.direction != Connector.UNDIRECTED:
            raise Exception('Other side of a relationship must be undirected.\n'
                            'Use - instead of <<: Node() << Relationship() - Node()')
        
        return Connector(a, b, Connector.LEFT)
        
    def sub(self, a, b):
        if a.type_ == 'node' and b.type_ == 'node':
            ghost_rel = Connector(
                AnonymousRelationship(), b, Connector.UNDIRECTED)
            return Connector(a, ghost_rel, Connector.UNDIRECTED)
        else:
            return Connector(a, b, Connector.UNDIRECTED)


class NodeLike(Element):
    type_ = 'node_like'
    comparator_factory = NodeLikeComparator


class Node(NodeLike):
    creation_counter = 1
    comparator_factory = NodeLikeComparator
    type_ = 'node'
    prefix = '('
    suffix = ')'
    
    def __init__(self, *labels, **props):
        self.idx = Node.creation_counter
        Node.creation_counter += 1
        self.variable = Text('%s_%d' % (self.type_, self.idx))
        self.labels = labels
        self.props = props
    
    def render(self, **ctx):
        parts = []
        if self.variable:
            parts.append(self.variable.render(**ctx))
        if self.labels:
            parts.append(':' + ':'.join(self.labels))
        if self.props:
            if parts:
                parts.append(' ')
            parts.append(repr(self.props))  # don't do this in real life, yo
        return ''.join((self.prefix, ''.join(parts), self.suffix))
    
    def match_expression(self):
        return self
    
    def returnable(self):
        return self.variable


class Relationship(Node):
    type_ = 'relationship'
    prefix = '['
    suffix = ']'


class AnonymousRelationship(Relationship):
    def __init__(self, *args, **kwargs):
        super(AnonymousRelationship, self).__init__(*args, **kwargs)
        self.variable = Empty()


class Connector(NodeLike):
    UNDIRECTED = 0
    LEFT = 1
    RIGHT = 2
    
    type_ = 'connector'
    
    def __init__(self, left, right, direction):
        self.left = left
        self.right = right
        self.direction = direction
    
    def render(self, **ctx):
        connector = {
            self.UNDIRECTED: '-',
            self.LEFT: '<-',
            self.RIGHT: '->',
        }[self.direction]
        return ''.join([self.left.render(**ctx), connector, self.right.render(**ctx)])
    
    def match_expression(self):
        return self


class Match(Element):
    type_ = 'match'
    
    def __init__(self, *expressions):
        self.expressions = []
        for e in expressions:
            match_expression = e.match_expression()
            if not e:
                raise Exception('Match only accepts match expressions.')
            self.expressions.append(match_expression)
    
    def render(self, **ctx):
        return 'MATCH ' + ', '.join(e.render(**ctx) for e in self.expressions)


class Return(Element):
    type_ = 'return'
    
    def __init__(self, *returnables):
        self.returnables = []
        for r in returnables:
            returnable = r.returnable()
            if not returnable:
                raise Exception('Return only accepts match returnables.')
            self.returnables.append(returnable)
    
    def render(self, **ctx):
        return 'RETURN ' + ', '.join(r.render(**ctx) for r in self.returnables)


class Bind(object):
    def execute(self, query):
        """Execute a cypher query"""
        raise NotImplementedError

    
class Engine(Bind):
    def __init__(self, url):
        self.url = url
    
    


print Node('Ingredient', name='Soy Milk') - Relationship('EXCLUDED_FROM') >> Node('Ingredient', name='Milk')
print
print Node('Admin', 'User', 'Contact', name='Zach') - Relationship('ADMINISTERS') >> Node('Product', upc=8011)
print 

zach = Node('User', name='Zach')
brian = Node('User', name='Brian')
print Match(zach - brian)
print
print Return(zach, brian)
```
